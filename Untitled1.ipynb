## Divide and Conquer - Merge Sort


### Explanation
Merge Sort is a Divide and Conquer algorithm. It divides the array into two halves, sorts each half recursively, and then merges them.

```python
def merge_sort(arr):
    if len(arr) <= 1:
        return arr

    mid = len(arr)//2
    left = merge_sort(arr[:mid])
    right = merge_sort(arr[mid:])

    return merge(left, right)

def merge(left, right):
    result = []
    i = j = 0

    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1

    result.extend(left[i:])
    result.extend(right[j:])
    return result
```

```python
arr = [38, 27, 43, 3, 9, 82, 10]
sorted_arr = merge_sort(arr)

print("Original:", arr)
print("Sorted:", sorted_arr)
```

```python
import time
import random

arr = [random.randint(1, 1000) for _ in range(1000)]

start = time.time()
merge_sort(arr)
end = time.time()

print("Execution Time:", end - start)
```

### Time Complexity
O(n log n)

### Observation
Merge Sort performs efficiently even for large inputs due to its divide and conquer approach.

## Sorting Algorithms Comparison

### Explanation
In this section, we compare the performance of Merge Sort, Quick Sort, and Python's built-in sort based on execution time for different input sizes.

```python
def quick_sort(arr):
    if len(arr) <= 1:
        return arr

    pivot = arr[len(arr)//2]

    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]

    return quick_sort(left) + middle + quick_sort(right)
```

```python
import time
import random

sizes = [100, 500, 1000, 2000]
merge_times = []
quick_times = []
builtin_times = []

for size in sizes:
    arr = [random.randint(1, 10000) for _ in range(size)]

    # Merge Sort
    start = time.time()
    merge_sort(arr.copy())
    merge_times.append(time.time() - start)

    # Quick Sort
    start = time.time()
    quick_sort(arr.copy())
    quick_times.append(time.time() - start)

    # Built-in Sort
    start = time.time()
    sorted(arr.copy())
    builtin_times.append(time.time() - start)
```

```python
import matplotlib.pyplot as plt

plt.plot(sizes, merge_times, label="Merge Sort")
plt.plot(sizes, quick_times, label="Quick Sort")
plt.plot(sizes, builtin_times, label="Built-in Sort")

plt.xlabel("Input Size")
plt.ylabel("Execution Time")
plt.title("Sorting Algorithms Comparison")
plt.legend()
plt.show()
```

### Observation
The built-in sort performs the fastest as it is highly optimized. Merge Sort shows consistent performance, while Quick Sort may vary depending on pivot selection.

## Greedy Algorithms

### Explanation
Greedy algorithms make the locally optimal choice at each step with the hope of finding the global optimum solution. They are efficient but may not always give the best result for all problems.

```python
def fractional_knapsack(items, capacity):
    # items = (weight, value)
    items.sort(key=lambda x: x[1]/x[0], reverse=True)

    total_value = 0

    for weight, value in items:
        if capacity >= weight:
            capacity -= weight
            total_value += value
        else:
            total_value += value * (capacity/weight)
            break

    return total_value
```

```python
items = [(10, 60), (20, 100), (30, 120)]
capacity = 50

print("Maximum Value:", fractional_knapsack(items, capacity))
```

```python
def activity_selection(start, finish):
    activities = list(zip(start, finish))
    activities.sort(key=lambda x: x[1])

    selected = [activities[0]]

    for i in range(1, len(activities)):
        if activities[i][0] >= selected[-1][1]:
            selected.append(activities[i])

    return selected
```

```python
start = [1, 3, 0, 5, 8, 5]
finish = [2, 4, 6, 7, 9, 9]

result = activity_selection(start, finish)
print("Selected Activities:", result)
```

### Observation
Greedy algorithms provide fast solutions by making optimal local choices. Fractional Knapsack gives an optimal solution, while Activity Selection efficiently selects maximum non-overlapping activities.

## Dynamic Programming

### Explanation
Dynamic Programming is used to solve problems with overlapping subproblems and optimal substructure. It stores results of subproblems to avoid recomputation, improving efficiency.

```python
def knapsack(W, wt, val, n):
    dp = [[0]*(W+1) for _ in range(n+1)]

    for i in range(1, n+1):
        for w in range(W+1):
            if wt[i-1] <= w:
                dp[i][w] = max(val[i-1] + dp[i-1][w-wt[i-1]], dp[i-1][w])
            else:
                dp[i][w] = dp[i-1][w]

    return dp[n][W], dp
```

```python
wt = [10, 20, 30]
val = [60, 100, 120]
W = 50

max_value, dp_table = knapsack(W, wt, val, len(wt))

print("Maximum Value:", max_value)

print("\nDP Table:")
for row in dp_table:
    print(row)
```

```python
def fib_recursive(n):
    if n <= 1:
        return n
    return fib_recursive(n-1) + fib_recursive(n-2)
```

```python
def fib_dp(n):
    dp = [0]*(n+1)
    dp[1] = 1

    for i in range(2, n+1):
        dp[i] = dp[i-1] + dp[i-2]

    return dp[n]
```

```python
import time

n = 30

start = time.time()
fib_recursive(n)
rec_time = time.time() - start

start = time.time()
fib_dp(n)
dp_time = time.time() - start

print("Recursive Time:", rec_time)
print("DP Time:", dp_time)
```

### Observation
Dynamic Programming significantly reduces execution time by storing previous results. Recursive Fibonacci is slow due to repeated calculations, while DP solves it efficiently.

## Travelling Salesman Problem (TSP)

### Explanation
The Travelling Salesman Problem (TSP) aims to find the shortest possible route that visits each city once and returns to the starting point. It is an NP-hard problem and grows exponentially with input size.

```python
from itertools import permutations

def tsp(graph):
    n = len(graph)
    nodes = list(range(n))
    min_path = float('inf')

    for perm in permutations(nodes[1:]):
        cost = 0
        k = 0

        for j in perm:
            cost += graph[k][j]
            k = j

        cost += graph[k][0]
        min_path = min(min_path, cost)

    return min_path
```

```python
graph = [
    [0, 10, 15, 20],
    [10, 0, 35, 25],
    [15, 35, 0, 30],
    [20, 25, 30, 0]
]

print("Minimum Cost:", tsp(graph))
```

### Observation
The brute-force solution of TSP checks all possible permutations, resulting in exponential time complexity O(n!). It becomes impractical for large inputs.
